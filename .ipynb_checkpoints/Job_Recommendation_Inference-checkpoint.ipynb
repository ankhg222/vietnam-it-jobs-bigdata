```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 1. Load dữ liệu và embeddings
df = pd.read_csv('job_data.csv')
job_embeddings = np.load('job_embeddings.npy')
print("Đã load dữ liệu jobs:", df.shape)
print("Đã load job_embeddings:", job_embeddings.shape)
```

```python
# 2. Khởi tạo lại mô hình NLP
model_name = 'paraphrase-multilingual-MiniLM-L12-v2'
model = SentenceTransformer(model_name)
```

```python
# 3. Hàm gợi ý công việc
def recommend_jobs(user_profile, top_k=5):
    # Chuyển user_profile thành embedding
    user_embedding = model.encode([user_profile])
    
    # Tính cosine similarity
    similarities = cosine_similarity(user_embedding, job_embeddings)[0]
    
    # Lấy top_k jobs
    top_k = min(top_k, len(df))
    top_indices = np.argsort(similarities)[::-1][:top_k]
    
    # Tạo kết quả
    result = df.iloc[top_indices].copy()
    result['similarity_score'] = similarities[top_indices]
    
    # Cập nhật danh sách các cột cần hiển thị dựa trên file CSV thực tế
    display_columns = ['title_clean', 'company', 'skills_clean', 'salary_final_vnd', 'similarity_score']
    # Lọc ra những cột có tồn tại trong df
    display_columns = [col for col in display_columns if col in result.columns]
    return result[display_columns]
```

```python
# 4. Kiểm thử hệ thống
print("\n--- Test Tiếng Việt ---")
user_1 = "Tôi muốn làm lập trình web backend với Python, đã có kinh nghiệm dùng SQL"
print("User profile:", user_1)
display(recommend_jobs(user_1, top_k=3))

print("\n--- Test Tiếng Anh ---")
user_2 = "Looking for artificial intelligence and NLP roles"
print("User profile:", user_2)
display(recommend_jobs(user_2, top_k=3))
```
